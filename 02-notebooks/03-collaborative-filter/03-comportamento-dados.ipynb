## 2. Criação da flag de avaliação de livros do Harry Potter

```python
train_data['FLAG_BOOK_HP'] = np.where(train_data['Book-Title'].str.contains('Potter'), 1, 0)
```

```python
# list_of_hp_books = train_data\
#     .query('FLAG_BOOK_HP == 1')['Book-Title']\
#     .unique()\
#     .tolist()

list_of_hp_books = ['Harry Potter and the Chamber of Secrets (Book 2)',
                    'Harry Potter and the Prisoner of Azkaban (Book 3)',
                    'Harry Potter and the Goblet of Fire (Book 4)',
                    'Harry Potter and the Order of the Phoenix (Book 5)',
                    "Harry Potter and the Sorcerer's Stone (Book 1)"]

list_of_hp_books
```

```python
pd.set_option('max_colwidth', 800)

for book in list_of_hp_books:

    print(f'>>>>>>>>>>>>> Livro escolhido: {book}')

    # DataFrame temporário com os usuários que avaliaram o livro do Harry Potter

    df_temp = train_data\
        .query(f'`Book-Title` == "{book}"')

    # Lista dos usuários 

    list_users_temp = df_temp['User-ID'].unique().tolist()

    # 

    df_train_data_temp = train_data\
        .query(f'`User-ID`.isin({list_users_temp})')['Book-Title']\
        .value_counts()\
        .reset_index()\
        .query(f'`Book-Title` != "{book}"')\
        .head(20)

    display(df_train_data_temp)
```

```python
lst_temp = train_data\
    .query('FLAG_BOOK_HP == 1')['Book-Title']\
    .unique()\
    .tolist()

[book for book in lst_temp if 'Book 6' in book]
```
